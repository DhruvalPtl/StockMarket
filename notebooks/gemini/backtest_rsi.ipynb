```python
import pandas as pd
import pandas_ta as ta  # You might need to install this: pip install pandas_ta

# 1. LOAD YOUR DATA
# Replace 'your_nifty_data.csv' with your actual file name
print("Loading data... this might take a minute for 100MB files...")
try:
    df = pd.read_csv('your_nifty_data.csv') # Or pd.read_excel() if it's .xlsx
except:
    print("Error: File not found. Make sure the file name is correct.")
    exit()

# Ensure column names are standard (adjust these if your file has different names)
# Example: If your time column is 'Date', rename it or use it as index
df.rename(columns={'OPEN': 'Open', 'HIGH': 'High', 'LOW': 'Low', 'CLOSE': 'Close'}, inplace=True)

# 2. CALCULATE INDICATORS
# Calculate RSI (14 periods)
df['RSI'] = ta.rsi(df['Close'], length=14)

# 3. DEFINE STRATEGY (The Logic)
# Buy Signal: RSI crosses above 30 (Recovering from oversold)
# Sell Signal: RSI crosses below 70 (Falling from overbought)

df['Signal'] = 0
df.loc[df['RSI'] < 30, 'Signal'] = 1  # Potential Buy Zone
df.loc[df['RSI'] > 70, 'Signal'] = -1 # Potential Sell Zone

# 4. SIMULATE TRADES
capital = 100000
position = 0
log = []

for i in range(1, len(df)):
    price = df['Close'].iloc[i]
    rsi = df['RSI'].iloc[i]
    
    # SIMPLE LOGIC: 
    # If RSI > 75 (Extreme Overbought), we SELL (Buy Put logic)
    if rsi > 75 and position == 0:
        entry_price = price
        position = -1 # Short
        print(f"SELL at {price} | RSI: {rsi:.2f}")
        
    # Exit Short if RSI goes back to 50
    elif position == -1 and rsi < 50:
        profit = entry_price - price
        capital += profit
        position = 0
        print(f"EXIT SHORT at {price} | Profit: {profit:.2f}")

print(f"Final Capital: {capital}")
```
